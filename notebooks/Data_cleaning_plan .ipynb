## Day 5 - Data cleaning Plan 

```python
## Import Libraries
import pandas as pd
import numpy as np
```

```python
## load the dataset
customers = pd.read_csv(r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\data\raw\Customers.csv")

operations = pd.read_csv(r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\data\raw\product_operations.csv")
```

```python
## Missing values
print("CUSTOMERS DATASET")

missing_customers = customers.isnull().sum()

missing_customers = pd.DataFrame({
    "Column": missing_customers.index,
    "Missing Values": missing_customers.values,
    "Missing %": ((missing_customers.values / len(customers)) * 100).round(2)
})

print(missing_customers)
```

```python
## Duplicate records 
print("Duplicate Customer IDs")

customer_duplicates = customers.duplicated(subset="customer_id").sum()
print(customer_duplicates)

print()

print("Duplicate Record IDs")

record_duplicates = operations.duplicated(subset="record_id").sum()
print(record_duplicates)
```

```python
## Data TYpes
print(customers.dtypes)

print(operations.dtypes)
```

```python
## Check categories 
category_columns = [
    "plan_category",
    "payment_status",
    "subscription_status"
]

for col in category_columns:
    print("\n",col)
    print(operations[col].value_counts(dropna=False))
```

```python
## Cleaning Plan Rules 
cleaning_plan = pd.DataFrame({

"Column":[
"Missing Values",
"Duplicate Records",
"Date Columns",
"Discount Percent",
"Tax Amount",
"Final Amount Paid",
"Plan Category",
"Payment Status",
"Subscription Status"
],

"Issue":[
"Null values",
"Duplicate IDs",
"Wrong datatype",
"Invalid percentage",
"Negative values",
"Negative values",
"Inconsistent labels",
"Inconsistent labels",
"Inconsistent labels"
],

"Action":[
"Impute using mode/median",
"Keep first record and remove duplicates",
"Convert to datetime",
"Replace invalid values",
"Validate and correct",
"Validate and correct",
"Standardize category names",
"Standardize category names",
"Standardize category names"
]

})

cleaning_plan
```

```python
import os

os.makedirs(r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\docs", exist_ok=True)

cleaning_plan.to_csv(
    r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\docs\Cleaning_Plan_v1.csv",
    index=False
)

print("Cleaning Plan v1 saved successfully in the docs folder.")
```
