```python
## Import Libraries

import pandas as pd
import numpy as np
```

```python
## Load Datasets 

customers = pd.read_csv(r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\data\raw\Customers.csv")

operations = pd.read_csv(r"C:\Users\SVS\Desktop\Product_Adoption_customer_sucess\data\raw\product_operations.csv")
```

```python
## Detect Data Type Issues 
print("Customers Data Types")
print(customers.dtypes)
```

```python
print("Operations Data Types")
print(operations.dtypes)
```

```python
## Detect Numeric as text 
numeric_columns = [
    "discount_percent",
    "tax_amount",
    "final_amount_paid"
]

for col in numeric_columns:
    if operations[col].dtype == "object":
        print(f"{col} is stored as TEXT")
    else:
        print(f"{col} is Numeric")
```

```python
## Malformed Dates 
operations["subscription_start_date"] = pd.to_datetime(
    operations["subscription_start_date"],
    errors="coerce"
)

operations["subscription_end_date"] = pd.to_datetime(
    operations["subscription_end_date"],
    errors="coerce"
)
```

```python
## Invalid Dates
print("Invalid Start Dates")

print(
operations["subscription_start_date"].isna().sum()
)

print("Invalid End Dates")

print(
operations["subscription_end_date"].isna().sum()
)
```

```python
## Invalid Numeric Values
## Discount Per
invalid_discount = operations[
(operations["discount_percent"]<0) |
(operations["discount_percent"]>100)
]

print("Invalid Discount Records")
print(invalid_discount)
```

```python
## Tax Amount 
invalid_tax = operations[
operations["tax_amount"]<0
]

print("Invalid Tax Records")
print(invalid_tax)
```

```python
## Amount Paid 
invalid_amount = operations[
operations["final_amount_paid"]<0
]

print("Invalid Final Amount")

print(invalid_amount)
```

```python
## Inconsistent  Categorie 
print(operations["plan_category"].unique())
```

```python
## Payment Status 
print(operations["payment_status"].unique()
)
```

```python
## subscription Status 
print(operations["subscription_status"].unique())
```

```python
for col in [
"plan_category",
"payment_status",
"subscription_status"
]:

    print("\nChecking :",col)

    print(
    operations[col][
    operations[col]!=operations[col].str.strip()
    ]
    )
```

```python
## Captalization 
for col in [
"plan_category",
"payment_status",
"subscription_status"
]:

    print("\nUnique values in",col)

    print(
    operations[col].str.lower().unique()
    )
```

```python
## Profiling Report 
profiling_report = pd.DataFrame({

"Check":[

"Invalid Start Dates",

"Invalid End Dates",

"Invalid Discount",

"Invalid Tax",

"Invalid Final Amount"

],

"Count":[

operations["subscription_start_date"].isna().sum(),

operations["subscription_end_date"].isna().sum(),

len(invalid_discount),

len(invalid_tax),

len(invalid_amount)

]

})
```
